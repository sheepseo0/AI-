```python
#3의 배수 판단
num = int(input('정수를 입력하세요:'))

if num % 3 == 0 and (num % 5 == 0):
    print(num,"은(는) 3의 배수이자 5의 배수입니다.")
elif num % 3 == 0:
    print(num,"은(는) 3의 배수입니다.")
```

```python
#LAB 3-3 (3)
a = int(input('당신은 성인인가요?(성인이면 1, 미성년이면 0):'))

if a == 0:
    print("당신은 미성년자입니다.")
elif a == 1:
    m = int(input("결혼을 하셨나요(기혼 : 1, 미혼 : 0):"))
    if m == 1:
        print("당신은 결혼한 성인입니다.")
    elif m == 0:
        print("당신은 결혼하지 않은 성인입니다.")
```
